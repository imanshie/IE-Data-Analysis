```python
import pandas as pd
from kmodes.kprototypes import KPrototypes
import plotly.express as px
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/prof-apartida/data-exercises/main/Credit_risk.csv', sep=',')
df = df[['CreditScore', 'Geography', 'Gender', 'Age', 'EstimatedSalary', 'Balance']].copy()
df.head()
```

```python
kproto = KPrototypes(n_clusters=3, init = "random", n_init = 5, verbose=0)
clusters = kproto.fit_predict(df, categorical=[1,2])
df.insert(0,'Cluster',clusters)

df
```

```python
fig = px.scatter( 
    df, 
    x="Balance", 
    y="EstimatedSalary", 
    color="Cluster", 
    title="Cluster by K prototypes",
)
fig.show()
```
